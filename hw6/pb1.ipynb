```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import RobustScaler

path = "./ml-100k/"
columns = ["user_id", "item_id", "rating", "timestamp"]

train = pd.read_csv(path + "u1.base", sep="\t", names=columns)
test = pd.read_csv(path + "u1.test", sep="\t", names=columns)
```

```python
train.head(), test.head()
```

Normalize

```python
train_mean = train['rating'].mean()
train_std = train['rating'].std()
scaler = RobustScaler()
scaler.fit(train[['rating']])
train[['rating']] = scaler.transform(train[['rating']])
test[['rating']] = scaler.transform(test[['rating']])
```

Create user-item matrix, user_similarity

```python

user_item_matrix = train.pivot_table(index='user_id', columns='item_id', values='rating')
user_similarity = cosine_similarity(np.nan_to_num(user_item_matrix.values))
user_similarity = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
```

```python
user_item_matrix.head(10)
```

```python
user_similarity.head(10)
```

KNN prediction

```python
def predict_rating(user_id, item_id, k=10):
    if item_id not in user_item_matrix.columns:
        return np.nan  

    # Users who rated the item
    users_rated = user_item_matrix[item_id].dropna()
    
    if users_rated.empty:
        return np.nan  # no one rated

    # Similarity scores
    sim_scores = user_similarity.loc[user_id, users_rated.index]
    
    # Top-k similar users without oneself
    top_k_users = sim_scores.sort_values(ascending=False).iloc[1:k+1]  

    # Ratings of the item from top-k users
    top_k_ratings = user_item_matrix.loc[top_k_users.index, item_id]

    # Weighted average
    if top_k_ratings.isnull().all():
        return np.nan  # all top-k users haven't rated

    weights = top_k_users[top_k_ratings.notnull()]
    ratings = top_k_ratings.dropna()
    
    if weights.sum() == 0:
        return np.nan  # avoid division by 0
    
    return np.dot(weights.values, ratings.values) / weights.sum()

predictions = []
ground_truth = []

for _, row in test.iterrows():
    pred = predict_rating(row['user_id'], row['item_id'], k=10)
    if not np.isnan(pred):
        predictions.append(pred)
        ground_truth.append(row['rating'])

# Compute RMSE
rmse = np.sqrt(mean_squared_error(ground_truth, predictions))
print(f"RMSE on test set: {rmse:.4f}")
```
